```python
import pandas as pd
import os

# Den Pfad zur CSV-Datei definieren (angenommen, sie liegt im 'data'-Ordner deines Projekts)
csv_file_path = os.path.join(os.getcwd(), '..', 'Data', 'Dataset.csv')

# CSV-Datei einlesen
df = pd.read_csv(csv_file_path)

# Die ersten Zeilen des Datensatzes anzeigen
print(df.head())

# Spaltenüberschriften prüfen
print(df.columns)

```

```python
# Jahr aus 'Date Added To Xplore' extrahieren und eine neue Spalte hinzufügen
df['Publication Year'] = pd.to_datetime(df['Date Added To Xplore'], errors='coerce').dt.year

# Schlüsselwörter für KI im Software-Engineering
keywords = [
    'AI', 
    'Artificial Intelligence', 
    'Machine Learning', 
    'Deep Learning', 
    'Neural Networks',
    'Convolutional Neural Networks (CNN)', 
    'Recurrent Neural Networks (RNN)', 
    'Transformer', 
    'Natural Language Processing', 
    'Computer Vision', 
    'AI-driven Development', 
    'Automated Testing', 
    'Code Generation', 
    'AI Augmentation'
]
df_filtered = df[df['Abstract'].str.contains('|'.join(keywords), case=False, na=False)]

# Reduzierter Datensatz
columns_to_keep = ['Document Title', 'Authors', 'Publication Year', 'Abstract', 
                   'Article Citation Count', 'Mesh_Terms']
df_filtered = df_filtered[columns_to_keep]

# Überblick über fehlende Werte
print(df_filtered.isnull().sum())

# Entferne Zeilen mit fehlenden Abstracts
df_filtered = df_filtered.dropna(subset=['Abstract'])

# Gefilterte Daten speichern
output_file_path = os.path.join(os.getcwd(), 'Filtered_Dataset.csv')
df_filtered.to_csv(output_file_path, index=False)
print(f"Bereinigter Datensatz wurde gespeichert unter: {output_file_path}")

```
